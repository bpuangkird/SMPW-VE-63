### การแก้ปัญหา (Problem Solving)

ทุกการประกอบอาชีพจำเป็นต้องแก้ปัญหา

#### มห้ศจรรย์เลข 4 สำหรับความจำสั้น

ในปี 2000 Nelson Cowan ตีพิมพ์ผลงานทางด้านจิตวิทยา ความจำสั้น (short-term memory:STM) จดจำได้ 4 อย่าง

#### ขั้นตอนการแก้ปัญหาตามแบบโพลยา

1. เข้าใจปัญหา (Understand)
1. วางแผนการแก้ปัญหา (Plan)
1. ดำเนินการแก้ปัญหาเพื่อหาคำตอบหรือผลเฉลย (Solve)
1. ตรวจสอบย้อนกลับ (Check)

#### กฎสี่ประการสำหรับการเข้าใจปัญหา

1. ตารางของตัวเลข (Tables/Number Theory)
1. วาดรูปหรือเรขาคณิต (Graphs/Geometry)
1. สูตรสมการ (Formulas/Algebra)
1. คำบรรยายหรือพรรณา (Words)

#### การได้มาซึ่งบทสรุปเพื่อการวางแผนแก้ปัญหา

บทสรุป (Conclusion) ได้มาจากการให้เหตุผล (Reasoning) การให้เหตุผลมีสองรูปแบบ

1. การให้เหคุผลแบบอุปนัย (Inductive reasonings)
    * เหตุผลที่ได้เรียกว่า การคาดคะเน (Conjecture) หรือการเดา (Guess)
1. การให้เหตุผลแบบนิรนัย (Deductive reasonings)
    * กฎ (Law) เกิดจากการสังเกตุความเป็นไปในธรรมชาติหรือจากการทดลอง (Empiricism)
    * ทฤษฎี (Theory) เิด้จากการประมวลความรู้ (Knowledge) ต้องมีการพิสูจน์
    * กฎเกณฑ์ (Rule) เกิดจากการตั้งขึ้นมา

#### แบบจำลองทางคณิตศาสตร์

แบบจำลองทางคณิตศาสตร์ (Mathematical modeling) เป็นการเปลี่ยนจากปัญหาในธรรมชาติ
ให้อยู่ในรูปของคณิตศาสตร์

1. กำหนดเป้าหมาย หรือสิ่งที่ไม่รู้ (Unknown) เป็นฟังก์ชันเป้าประสงค์ (Objective function)
1. ร่างรูป (Sketch) แสดงความสัมพันธ์ระหว่างฟังก์ชันเป้าประสงค์และตัวแปรอิสระต่าง ๆ
1. สร้างสูตรสมการ (Formulating equations) จากรูปที่ร่างไว้
    * กำจัดตัวแปร จากความสัมพันธ์ที่คงที่
    * กำหนดขอบเขตของปัญหา (Domain or Boundary of the problem)
1. ดำเนินการแก้ปัญหาเพื่อหาสิ่งที่ไม่รู้ ให้เป็นผลเฉลย (Solution)

## ฟิสิกส์และกลศาสตร์

ฟิสิกส์ (Physics) เป็นวิชาที่ศึกษาเกี่ยวความเป็นไปในธรรมชาติ ในแง่ของมวล (Mass) แรง (Force) และพลังงาน (Energy)
โดยทำการวัดปริมาณที่วัดได้ (Measurable quantity) ปริมาณที่วัดได้วัดโดยการเทียบกับหน่วยการวัด (Unit) สิ่งที่ได้จากการวัดจำเป็นต้องบอกทั้ง ปริมาณ (Quantity) หน่วย (Unit) และความผิดพลาดในการวัด (Error) 


ความผิดพลาดในการวัดมักบอกโดยอ้อมจากเลขนัยสำคัญ (Significant figure) ตัวสุดท้ายเป็นค่าที่ไม่แน่นอน ในเชิงวิศวกรรมศาสตร์ เครื่องมือวัดที่ดีสุด วัดได้แม่นยำที่สุดได้ไม่เกิน 1% นั่นคือปริมาณที่ได้จากการวัดถูกรายงานด้วยเลขนัยสำคัญ
ไม่เกิน 3 ตัว

กลศาสตร์ เป็นสาขาย่อยของฟิสืกส์ศึกษา มวล แรง และการเคลื่อนที่ โดยมีปริมาณการวัดพื้นฐานอยู่ 4 อย่างคือ

1. ความยาว (Length)
1. เวลา (Time)
1. มวล (Mass)
1. แรง (Force)

กลศาสตร์แบ่งตามการเปลี่ยนรูปของวัตถุเมื่อมีแรงมากระทำ

1. กลศาสตร์ของวัตถุแข็งเกร็ง (Rigid body mechanics) แบ่งเป็น
    * สถิตศาสตร์ (Statics)
    * พลศาสตร์ (Dynamics)
1. กลศาสตร์ของวัตถุที่เปลี่ยนรูปได้ (Deformable body mechanics)
    * กลศาสตร์ของวัสดุ (Mechanics of Materials)
1. กลศาสตร์ของของไหล (Fluid mechanics) วัตถุเปลี่ยนรูปโดยไม่ต้องมีแรงภายนอกมากระทำ

## สสาร (Matters)

> **เอกสารอ้างอิง** John Bird _Science and Mathematics for Engineering_ 6th Edition, Routledge, 2020.


ก่อนหน้านี้มีความคิดว่าสสาร (Matter) คือสิ่งที่ต้องการที่อยู่และมีน้ำหนัก และใช้เวลาสักครู่ที่นักวิทยาศาสตร์จะรู้ว่าอากาศและแก๊สก็เป็นสสารด้วย

นิยามทางวิทยาศาสตร์ของสสารคือสิ่งที่ต้องการที่อยู่ (Space) และมี__มวล__(mass)

มีการพบว่าสสารมีความไม่ต่อเนื่อง (not continuous) ทำให้เกิดทฤษฎีโมเลกุลของสสาร (Mplecular theory of matters) และทฤษฎีอะตอมของสสาร (Atomic theory of matter) เกิดขึ้น โดยบอกว่าสสารประกอบไปด้วยอนุภาคขนาดเล็กเรียกว่าโมเลกุล (Molecules) และอะตอม (Atoms) โดยเคยมีความคิดว่าสสารประกอบด้วยอะตอมเป็นหน่วยที่เล็กที่สุดและไม่สามารถแบ่งแยกอะตอมได้ แต่จากการทดลองได้แสดงว่าอะตอมประกอบด้วยอนุภาคที่เล็กกว่าอาทิ โปรตอน (Proton) นิวตรอน (Neutron) และอิเล็กตรอน (Electrons) 

พืจารณาถึงความแบ่งแยกต่อไปไม่ได้ พบว่าโปรตรอน และนิวตรอนประกอบด้วยอนุภาคที่แบ่งแยกต่อไปไปไม่ได้เรียกว่าควาก (Quarks) นอกจากนี้ยังมีการค้นพบอนุภาคที่ไม่เป็นส่วนหนึ่งของอะตอมอีกด้วย
