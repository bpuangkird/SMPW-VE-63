# กลศาสตร์ของของไหล

ปริมาณพื้นฐานเพิ่มเติมสำหรับกลศาสตร์ของไหล

1. ความดัน (Pressure)
1. ความเร็ว (Velocity)

## การทดลองน้ำหยุดนิ่ง

น้ำและมีน้ำแข็งลอยอยู่ โดยที่ระดับน้ำเต็มแก้ว โดยมีน้ำแข็งลอยมีส่วนเหนือผิวน้ำ สังเกตุดูว่าเมื่อน้ำแข็งละลายน้ำจะล้นแก้วไหลลงมานองพื้นหรือไม่

## ความดัน

ความดันกระทำต่อพื้นผิว ให้คำจำกัดความว่าเป็นแรงที่ตั้งฉากกับพื้นผิวเรียบต่อหน่วยของพื้นที่ของพื้นผิว ในหน่วยมาตรฐาน $\rm{SI}$ ความดันมีหน่วยเป็นปาสคาล ($\rm{pascal}$ ตัวย่อคือ $\rm{Pa}$) เทียบเท่ากับแรง 1 นิวตันกระทำต่อพื้นผิวหนึ่งตารางเมตร

เชียนในรูปสูตรสมการ

$$p = \dfrac{F}{A} \quad \rm{pascals}$$

โดยที่ $F$ คือแรงในหน่วยนิวตันกระทำอย่างสม่ำเสมอทั่วพื้นที่ และทำมุมฉากกับพื้นที่ $A$ ตารางเมตร

## ความดันของของไหล

ของไหลอาจเป็นไปได้ทั้งของเหลวหรือแก๊สและมีองค์ประกอบพื้นฐาน 4 ประการที่ควบคุมความดันภายในของไหลคือ

1. ความดันที่ระดับความลึกหนึ่งในของไหลชนิดหนึ่งมีค่าเท่ากันในทุกทิศทาง
1. ความดันที่ระดับความลึกหนึ่งในของไหลชนิดหนึ่ง ไม่ขึ้นกับรูปร่างของภาชนะที่บรรจุของไหลนั้น
1. ความดันกระทำในทิศตั้งฉากกับผิวของภาชนะบรรจุ
1. เมื่อมีความดันมากระทำกับของไหล ความดันจะถูกส่งผ่านไปในของไหลในลักษณะที่เท่ากันทุกทิศทาง

ความดัน $p$ ในของไหล ณ จุด ๆ หนึ่งขึ้นกับองค์ประกอบ 3 ประการ

1. ความหนาแน่นของของไหล $\rho$ $\rm{\;kg/m^3}$
1. น้ำหนักของของไหล ซึ่งขึ้นกับค่าความเร่งจากแรงโน้มถ่วงของโลก $g$ มีค่า $\;9.81\;m/s^2$ หรือในหน่วย $\rm{N/kg}$
1. ความสูงของของไหล $h$ เมตรที่อยู่เหนือจุดขึ้นไป

$$p = \rho g h \quad \rm{pascals}$$
